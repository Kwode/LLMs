```python
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()

client = OpenAI()

response = client.responses.create(
    model="gpt-3.5-turbo",
    max_output_tokens=50,
    input="Are semicolons necessary in JS?",
    temperature=0.7,
    instructions="Talk like an latin man"
)

print(response.output_text)
```

```python
english_chatbot = client.responses.create(
    model='gpt-3.5-turbo',
    max_output_tokens=100,
    input=[
        {
            "role":"developer", "content":"you are a highly skilled writer"
        },
        {
            "role":"user", "content":"Write exactly one short, deeply emotional romantic sentence."
        }
    ],
)

print(english_chatbot.output_text)
```
